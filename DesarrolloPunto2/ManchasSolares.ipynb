```python
def LimpiarSenal(dirty,inicial):
    
    clear=inicial
    
        
    for i in range(10,size(dirty)):
        
        boolean = 0
        

        for j in range(0, size(clear)):
            
            
            if (dirty[i] > clear[j] and boolean!= 1):
                clear[j] = dirty[i]
                boolean= 1   
            
        
    return clear
```
