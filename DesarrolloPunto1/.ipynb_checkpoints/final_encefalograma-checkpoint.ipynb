```python
#Arreglo con las normas al cuadrado de los coeficientes

coef_nor_2=np.empty([24,400])

for i in range(len(fft_coef)):
    
    for j in range(400):
    
        norma_i_j=power(abs(((fft_coef[i])[j])),2)
        (coef_nor_2[i])[j]=norma_i_j

print coef_nor_2
```

```python
#Función para limpiar la señal: dejar sólo los coeficientes  con normas al cuadrado 

def LimpiarSenal(dirty):
    
    n=len(dirty)
    clear=np.empty(n)
        
    for i in dirty:
        
        cuenta=0
        
        for j in dirty:
            
            if i>=j:
                
                cuenta=cuenta+1
        
        for k in range(10):
        
            if cuenta==(n-k):
            
                clear[k]=i
    
    for i in range(n-10):
        
        clear[10+i]=0
            
            
```

```python
prueba=LimpiarSenal(coef_nor_2[1])
print prueba
```
