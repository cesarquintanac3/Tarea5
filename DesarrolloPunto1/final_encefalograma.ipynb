```python
#Función para limpiar la señal: dejar sólo los coeficientes  con normas al cuadrado 

def LimpiarSenal(dirty):
    
    n=len(dirty)  
    fin_clear=[]
    
    clear=(-dirty).argsort()[0:10]
    
    for i in range(size(clear)):
        
        k=clear[i]
        fin_clear.append(dirty[k])
        

    for j in range(10,n):
        
        fin_clear.append(0.0)
            
    return fin_clear    
```

```python
#Limpiar señal

sen_limpia=np.empty([24,400])

for i in range (len(coef_nor_2)):
    sen_limpia[i]=LimpiarSenal(coef_nor_2[i])
    
print sen_limpia

for i in range(24):
    plot(fft_freq,sen_limpia[i])
```
