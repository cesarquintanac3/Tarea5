```python
def guardardatos(dato):

    #Arreglo que guarda los arreglos con la información de cada electrodo
    electrodos=np.empty([24,len(dato)])

    #Ciclo para guardar la información en cada arreglo
        
    for i in range(24):
            
        for j in range(len(data)):

                (electrodos[i])[j]=(data[j])[i]

    return electrodos
```

```python
electrodo=guardardatos(data)
print electrodo
```

```python
from scipy.fftpack import fft, fftfreq

fft_coef=[]
n=400

for i in electrodo:
    
    fft_i=fft(i)/n
    fft_coef.append(fft_i)

print fft_coef
```
